```python
import pymysql
import csv

# MySQL 서버 접속 정보
db = pymysql.connect(
  host="hobtdb.cgvif8lsqknk.us-east-2.rds.amazonaws.com",
  port=3306,
  user="admin",
  password="1234qwer",
  database="hobt",
  charset='utf8mb4',
  cursorclass=pymysql.cursors.DictCursor
)

# CSV 파일 읽기
with open('HoBT_web.csv', newline='', encoding='utf-8') as csvfile:
    reader = csv.DictReader(csvfile)
    for row in reader:
        # MySQL에 데이터 삽입
        cursor = db.cursor()
        sql = "INSERT INTO test_questions (answer, similar_answer, question, test_year, sub_category, main_category, etc) VALUES (%s, %s, %s, %s, %s, %s, %s)"
        values = (row['answer'], row['similar_answer'], row['question'], row['test_year'], row['sub_category'], row['main_category'], row['etc'])
        cursor.execute(sql, values)
        db.commit()
        print(cursor.rowcount, "record inserted.")
```
